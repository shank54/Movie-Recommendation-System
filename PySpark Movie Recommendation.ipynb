#### Q) Report of movies split title and year

```python
movies_latest_pd['movie_title'] = "movie_title"
movies_latest_pd['year'] = 0
```

```python
for t in range(len(movies_latest_pd['title'])):
    our_string = movies_latest_pd['title'][t].rstrip()
    movie_name = our_string[:-6]
    year = our_string[-5:-1]
    movies_latest_pd['movie_title'].values[t] = movie_name
    if year.isdigit():
        movies_latest_pd['year'].values[t] = year
```

```python
columns = ['title']
movies_latest_pd = movies_latest_pd.drop(columns, axis=1)
```

#### Q) movies by genres

```python
movieId = []
genres = []
movie_title = []
year = []
for i,row in movies_latest_pd.iterrows():
    movie = row.movieId
    movie_nam = row.movie_title
    yea = row.year
    for genre in row.genres.split("|"):
        movieId.append(movie)
        genres.append(genre)
        movie_title.append(movie_nam)
        year.append(yea)

movies_split_genres = pd.DataFrame({"movieId":movieId,"genres":genres,"movie_title":movie_title,"year":year})
```

#### Machine Learning model to recommend movies

```python
from pyspark.ml.recommendation import ALS
from pyspark.ml.evaluation import RegressionEvaluator
```

```python
df_data_2.show()
```

```python
df_data_2.describe().show()
```

```python
training,test = df_data_2.randomSplit([0.7,0.3])
```

```python
als = ALS(maxIter=5,regParam=0.01,userCol='userId',itemCol='movieId',ratingCol='rating',coldStartStrategy="drop")
```

```python
model = als.fit(training)
```

```python
predictions = model.transform(test)
```

```python
predictions.show()
```

```python
evaluator = RegressionEvaluator(metricName='rmse',labelCol='rating',predictionCol='prediction')
```

```python
rmse = evaluator.evaluate(predictions)
```

```python
print(rmse)
```
